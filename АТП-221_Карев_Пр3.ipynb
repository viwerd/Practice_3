```python
import numpy as np

a = np.array([[[1,2,3],[4,5,6],[6,7,8],[8,9,10]]])
print(a) # Вывод самого массива
print(a.ndim) # Измерение массива
print(a.shape) # Состав массива , 1 - 3-х мерный , 4 строки , 3 столбца
print(a.size) # Кол-во элементов массива
print(a.itemsize) # Кол-во памяти  1 элемента массива
print(a.nbytes) # Кол-во памяти занимаемое всем массивом
#print(a[1, 3, 1]) # Узнаём что стоит на этом месте в массиве ( 1 строка , 3 место) P.S. НУМЕРАЦИЯ С НУЛЯ
#print(a[1, :]) # Выводит все элементы 2 строки

b = np.zeros((2, 2)) # Создаёт массив заполненный нулями , ones , full , random.randint , identity
print(b)
```

```python
print("1) массив чисел от 0 до 9:")
arr = np.arange(10)
arr
```

```python
print("2) массив чисел(3х3), состоящий из единиц и f.ones:")
arr = np.ones((3, 3))
arr
```

```python
print("3) массив чисел(2х2), состоящий из рандомов:")
arr = np.random.randint(1, 11, size=(2, 2))
arr
```

```python
arr = np.random.rand(5, 5)
print("4.1) массив 5х5")
arr
```

```python
print("4.2) массив 5х5 со средним значением:")
srednee = np.mean(arr)
srednee # Output медианы
```

```python
print("5) массив 5х5 с рандомным заполнением от 0 до 1:")
arr = np.random.rand(5, 5)
print("Сам массив: " , arr)
col_sums = arr.sum(axis=0)
print("Суммы столбцов = " , col_sums)
```
